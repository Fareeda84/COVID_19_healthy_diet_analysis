## Predicting confirmed cases based on the diet by building a Machine Learning model

```python
data
```

```python
#Creating new dataFrame by extracting diet columns from the original data
data_diet = data.iloc[:, 1:24]
data_diet.columns
```

```python
# Check whether data have null values 
data_diet.isna().sum()
```

```python
data_diet
```

```python
# Split data into x and y
np.random.seed(42)

x = data_diet
y = data["Confirmed"]
x.shape, y.shape
```

```python
import sklearn

# Split tarin and test data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# Check x_train, y_train, x_test, y_test shape
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Build Machine learning model using RandonForestRegressor model

from sklearn.ensemble import RandomForestRegressor

# Build model
model = RandomForestRegressor()
# Fit the model
model.fit(x_train, y_train)
```

```python
# Check model score
model.score(x_test, y_test)
```

```python
# Building another model (Ridge Regression model)
from sklearn.linear_model import Ridge

# Fitting Ridge model
model1 = Ridge()
model1.fit(x_train, y_train)

# Check the Ridge model score on test data
model1.score(x_test, y_test)
```

## Model evaluation metrics

**R^2 (coefficient of determination)**

```python
# R^2
from sklearn.metrics import r2_score

# Fill an array with y_test mean
y_test_mean = np.full(len(y_test), y_test.mean())
y_test_mean.shape, y_test.shape
```

```python
# Check r2_score 
r2_score(y_test, y_test_mean)
```

**Mean absolute error** 

```python
# Absolute mean error
from sklearn.metrics import mean_absolute_error

# model prediction values
y_preds = model.predict(x_test)
# Absolute mean error
mae = mean_absolute_error(y_test, y_preds)
mae
```

```python
# DataFrame with actual and predicted values and their differences
Data_frame = pd.DataFrame({"Actual_values" : y_test,
                     "Predicted_values" : y_preds,
                      "Difference" : y_test - y_preds})
Data_frame
```

**Mean square error**

```python
# Mean square error
from sklearn.metrics import mean_squared_error

# Model prediction values
y_preds = model.predict(x_test)
# mean squared error
mse = mean_squared_error(y_test, y_preds)
mse
```

```python
# Calculating mean squared error by hand
squared = np.square(Data_frame["Difference"])
squared.mean()
```
